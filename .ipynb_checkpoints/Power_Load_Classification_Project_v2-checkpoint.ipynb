# Power System Load Type Classification

**Objective:**
Develop a machine learning model to predict the 'Load_Type' (Light_Load, Medium_Load, Maximum_Load) based on historical power system data.

**Steps:**
1. Data Loading & Preview
2. Exploratory Data Analysis (EDA)
3. Data Preprocessing
4. Feature Engineering
5. Model Selection & Training
6. Model Evaluation (using last month as test set)
7. Conclusion & Output


```python
# 1. Data Loading & Preview
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Load the data
data = pd.read_csv('c:/Users/Varun TD/Desktop/Use Ready Assignment 2/data/load_data.csv')
data.head()
```

```python
print('Data shape:', data.shape)
print('Columns:', data.columns.tolist())
print(data.info())
print(data.describe())

# Check for missing values
print('Missing values per column:')
print(data.isnull().sum())

# Distribution of Load_Type
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x='Load_Type', data=data)
plt.title('Load_Type Distribution')
plt.show()

# Correlation heatmap
import numpy as np
plt.figure(figsize=(10,6))
sns.heatmap(data.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation')
plt.show()
```

```python
# 3. Data Preprocessing
# Convert 'Date_Time' to datetime
data['Date_Time'] = pd.to_datetime(data['Date_Time'], errors='coerce')

# Encode Load_Type
le = LabelEncoder()
data['Load_Type_Encoded'] = le.fit_transform(data['Load_Type'])

# Feature selection
features = ['Usage_kWh', 'Lagging_Current_Reactive.Power_kVarh', 'Leading_Current_Reactive_Power_kVarh', 'CO2(tCO2)', 'NSM']
X = data[features]
y = data['Load_Type_Encoded']

# Handle missing values (simple imputation)
X = X.fillna(X.mean())

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 4. Train/Test Split (last month as test set)
data_sorted = data.sort_values('Date_Time')
last_month = data_sorted['Date_Time'].dt.month.max()
last_year = data_sorted['Date_Time'].dt.year.max()
test_mask = (data_sorted['Date_Time'].dt.month == last_month) & (data_sorted['Date_Time'].dt.year == last_year)

X_train = X_scaled[~test_mask]
y_train = y[~test_mask]
X_test = X_scaled[test_mask]
y_test = y[test_mask]

print(f'Train set: {X_train.shape[0]} samples, Test set: {X_test.shape[0]} samples')
```

```python
# 5. Model Selection & Training
clf = RandomForestClassifier(random_state=42, n_estimators=100)
clf.fit(X_train, y_train)

# 6. Model Evaluation
y_pred = clf.predict(X_test)
print('Classification Report:')
print(classification_report(y_test, y_pred, target_names=le.classes_))

print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))

print('Accuracy:', accuracy_score(y_test, y_pred))
```

## Conclusion
- Developed a Random Forest Classifier to predict Load_Type.
- Used the last month as a test set for realistic evaluation.
- Evaluated using accuracy, precision, recall, and F1-score.
